# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importa o dataset para um DataFrame Pandas

```python
df = pd.read_csv('ixps_global_geocoded.csv')
```

# Analise do numero de IXPs por pais
## Plota um grafico de barras para visualizar a distribuicao de IXPs por pais

```python
plt.figure(figsize=(15, 6))
df['country'].value_counts().plot(kind='bar')
plt.title('Numero de IXPs por Pais')
plt.xlabel('Pais')
plt.ylabel('Numero de IXPs')
plt.xticks(rotation=90)
plt.show()
```

# Analise da distribuicao de IXPs por regiao
## Plota um grafico de pizza para visualizar a proporcao de IXPs em cada regiao

```python
plt.figure(figsize=(8, 8))
df['region'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribuicao de IXPs por Regiao')
plt.axis('equal')
plt.show()
```

# Analise da distribuicao de IXPs por ano de criacao
## Plota um histograma para visualizar a distribuicao de IXPs por ano de criacao

```python
plt.figure(figsize=(10, 6))
plt.hist(df['year'], bins=15)
plt.title('Histograma de IXPs por Ano de Criacao')
plt.xlabel('Ano de Criacao')
plt.ylabel('Numero de IXPs')
plt.show()
```

# Analise da distribuicao de IXPs por status
## Plota um grafico de barras para visualizar a distribuicao de IXPs por status

```python
plt.figure(figsize=(10, 6))
df['status_pch'].value_counts().plot(kind='bar')
plt.title('Distribuicao de IXPs por Status')
plt.xlabel('Status')
plt.ylabel('Numero de IXPs')
plt.show()
```

# Analise da distribuicao de IXPs por LMIC
## Plota um grafico de barras para visualizar a distribuicao de IXPs por LMIC

```python
plt.figure(figsize=(10, 6))
df['lmic'].value_counts().plot(kind='bar')
plt.title('Distribuicao de IXPs por LMIC')
plt.xlabel('LMIC')
plt.ylabel('Numero de IXPs')
plt.show()
```

# Analise da distribuicao geografica de IXPs
## Plota um mapa de dispersao para visualizar a distribuicao geografica de IXPs


```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Lon'], df['Lat'])
plt.title('Mapa de Dispersao de IXPs')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
